```python
from GeoJSONExporters import GeoJSONPointExporter
from GeoJSONExporters import GeoJSONPlaceExporter
import osmnx as ox
```

### 1. Export given a pair (latitude, longitude), and a radius

```python
latitude = 18.928184
longitude = -99.210153
radius = 1000

geoJSONExporter = GeoJSONPointExporter(latitude, longitude, radius)

geoJSONExporter.drawGraph("Figures/map_1.png")
```

```python
dataFrame = geoJSONExporter.exportToGeoJSON("Data/sidewalks_point.geojson")
```

### 2. Export given a place (Monterrey for instance)

```python
city = "Monterrey"
state = "Nuevo Leon"
country = "México"

placeExporter = GeoJSONPlaceExporter(city, state, country)
placeExporter.drawGraph("Figures/monterrey.png")
```

```python
dataFrame = placeExporter.exportToGeoJSON("Data/sidewalks_monterrey.geojson")
```
